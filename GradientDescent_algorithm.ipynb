```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
def gadieint_descent(x,y):
  m=b=0
  n=len(x)
  iterations=1000
  Learning_rate=0.001
  for i in range(iterations):
    y_predicted=m*x+b
    #cost function is also getting updated 
    cost_function=1/n*(sum(y-y_predicted))
    #m_derivative=-2/n*(sum(x*(y_predicted)))
    #n= length of independent variables in this case it is x 
    m_derivative=-2/n*(sum(x*(y-y_predicted)))
    b_deriavtive=-2/n*(sum(y-y_predicted))
    #updation of m and b 
    #m_new = m_old - Learning_ rate*m_derivative
    #b_new=b-old-Learning_rate*b_deriavtive
    # but since we are updating we donot need to craete separate m_new ,b_new 
    m = m - Learning_rate*m_derivative
    b = b - Learning_rate*b_deriavtive
    print("m={} ,b={} ,i={} ,cost_function={} ".format(m,b,i,cost_function))
```

```python
gadieint_descent(x,y)
```
