```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math
```

```python
sns.set()
```

```python
pi = math.pi
```

```python
x = np.linspace(0, 3*pi, 50)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
z = np.cos(x)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(x, z, c = "r", marker = "*", linestyle = "-.", label = "cos curve")
plt.scatter(x, y, color = "b", marker = "^", s = 100, label = "sin curve")
plt.title("sin and cos viz.", fontsize = 20)
plt.xlabel("redian values")
plt.ylabel("sin and cos")
plt.xticks(range(0, 10))
#plt.yticks([-1, 0, +1])
plt.grid()
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x, z, c = "r", marker = "*", linestyle = "-.", label = "cos curve")
#ax1.title or other customizations
ax2.scatter(x, y, color = "b", marker = "^", s = 100, label = "sin curve")
plt.show()
```
